```python
from __future__ import division

import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
import numpy as np
from sklearn.decomposition import PCA
```

```python
data = np.array([[x, np.random.uniform(0.9, 1.1)*2*x] for x in range(1, 10)])
pca = PCA(n_components=2)
pca.fit(data)
print pca.transform(data)
```

```python
[type(t) for t in zip(pca.components_)]
```

```python
mins = 0.1*pca.components_
maxs = 1.1*pca.components_
print mins
plt.plot([mins[0,0], maxs[0,0]], [mins[0,1], maxs[0,1]])
X, Y = zip(*data)
plt.plot(X, Y, 'o')
```
